```python
table = pd.read_csv("Cricket.csv")
```

```python
table = table.astype(str)
table
```

```python
# prompt: Using dataframe table: find the player with most runs

# Convert 'Runs' column to numeric, handling potential errors
table['Runs'] = pd.to_numeric(table['Runs'], errors='coerce')

# Find the player with the maximum runs
player_with_most_runs = table.loc[table['Runs'].idxmax()]

# Print the player's name and runs
print(f"The player with the most runs is {player_with_most_runs['Player']} with {player_with_most_runs['Runs']} runs.")
```
